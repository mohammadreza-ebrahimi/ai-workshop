# Merge and Concatenation

```python
df1 = pd.DataFrame({'A':np.arange(15,18),'B':np.arange(32,35),'C':np.arange(25,28)})
df1
```

```python
df2 = pd.DataFrame({'A':np.arange(10,14),'B':np.arange(41,45),'C':np.arange(18,22)})
df2
```

```python
# to merge these two:
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df = pd.concat([df1, df2], ignore_index=False, keys=['First', 'Second'])
df  # now it has two indexes
```

```python
df3 = pd.DataFrame({
    'Time': ['H1', 'H2', 'H3', 'H4'],
    'Temp': ['14', '18', '22', '21'],
    'Color': ['B', 'B', 'R', 'Y']
})
df3
```

```python
df4 = pd.DataFrame({
    'Time': ['H1', 'H2', 'H3', 'H4'],
    'Wind Speed': ['20', '26', '30', '33'],
    'Loudness': ['L', 'M', 'H', 'H']
})
df4
```

```python
# as  you can see the two data frames abow has one comon column.
res = pd.merge(df3, df4, on='Time')
res
```

```python
df5 = pd.DataFrame({
    'Time': ['H1', 'H2', 'H3', 'H4', 'H5'],
    'Temp': ['14', '18', '22', '21', '19'],
    'Color': ['B', 'B', 'R', 'Y', 'Y']
})
df5
```

```python
df6 = pd.DataFrame({
    'Time': ['H1', 'H2', 'H3', 'H4', 'H6'],
    'Wind Speed': ['20', '26', '30', '33', '25'],
    'Loudness': ['L', 'M', 'H', 'H', 'M']
})
df6
```

```python
# there are two ways to merge these two
# first way: just the things in comon. will throw away differences
result = pd.merge(df5, df6, on='Time', how="inner")
result
```

```python
# Second way: all the things exists. but NaN for missed information
result = pd.merge(df5, df6, on='Time', how='outer')
result
```
